### **1-3. 생성자**
* 클래스 인스턴스가 생성될 때 호출함
* __ init __(self, 변수1, 변수2..):
* self 매개변수는 항상 첫번째 오며, 자기 자신을 가리킴
* 이름이 꼭 self일 필요는 없지만, 관례적으로 self를 사용
* 생성자에서는 해당 클래스가 다루는 데이터를 정의

````python
class Dog:
  def __init__(self):
    print(self, '만들어짐')
    self.name = '이름없음'
    self.age = 0
````

````python
Rucy = Dog()
print(Rucy.name)
print(Rucy.age)


PPomi = Dog()
print(PPomi.name)
print(PPomi.age)
````

````python
class Dog:
  def __init__(self, name, age, family = '포메'):     # self는 변수
    self.name = name   # 변수들
    self.age = age
    self.family = family
````

````python
Rucy = Dog('루시', 12)
print(Rucy.name, Rucy.age, Rucy.family)
````

````python
PPomi = Dog('뽀미', 6, '폼피츠')
print(PPomi.name, PPomi.age, PPomi.family)
````

### **1-4. 메소드 정의하기**

* 멤버함수라고도 하며, 해당 클래스의 객체에서만 호출가능한 함수
* 메소드는 객체에서만 호출되며, 해당 객체의 속성에 대한 연산을 행함
* 객체이름.메소드명() 형태로 호출됨

````python
class Counter:
  def __init__(self):
    self.num = 0
  def increment(self):
    self.num += 1
  def current_value(self):
    return '대기인원 : ' +str(self.num)
  def reset(self):
    self.num = 0
````

````python
KBbank = Counter()
print(KBbank.num)

KBbank.increment()
KBbank.increment()
KBbank.increment()
print(KBbank.num)

KBbank.current_value()
````

````python
HanaBank = Counter()
print(HanaBank.num)

HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
print(HanaBank.num)
HanaBank.current_value()
HanaBank.reset()
print(HanaBank.num)
````

### **1-5. 메소드 타입**
* instance method : 객체 형태로 호출되기 때문에 해당 메소드를 호출한 객체에만 영향을 미침
* class method : class로 호출(함수 선언 위에 @staticmethod 라고 표기)

````python
class Math:
  def add(self, x,y):
    return x + y
  def multiply(self, x, y):
    return x * y
    
````

````python
math = Math()
result1 = math.add(10, 5)
result2 = math.multiply(10,5)
print(result1, result2)
````

````python
class Math:
  @staticmethod # 객체를 생성하지 않아도된다. (미리 메모리에 올라가므로 클래스이름만 적으면 된다.)
  def add(x,y):
    return x + y
  @staticmethod
  def multiply(x, y):
    return x * y
````

````python
result1 = Math.add(10, 5)
result2 = Math.multiply(10, 5)
print(result1, result2)
````
