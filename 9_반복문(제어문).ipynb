# **1. 대입 연산자**

* =, +=, -=, *=, /=, %=
* 왼쪽에 저장된 값과 오른쪽의 값을 연산한 뒤 다시 왼쪽에 대입

````python
num = 10

num += 1 # num = num + 1
print(num)

num -= 1 # num = num - 1
print(num)

num*= 10 # num = num * 10
print(num)

num/= 10 # num = num / 10
print(num)

num %= 2 # num = num % 2
print(num)
````

# **2. while 문**

```
while 조건식:
  조건식이 True인 동안 반복할 문장
  ...

```

````python

i = 1
while i <= 5:
  print('Hello Python')
  i += 1

print('현재 i의 값 : {}'.format(i))
````

````python
# 1 ~ 10 까지의 합

i = 1
sum = 0 # 1~10까지 합을 저장할 변수

while  i <= 10:
  sum += i
  i += 1


print('1부터 10까지의 합 : {}'.format(sum))
print(f'1부터 10까지의 합 : {sum}')
````

# **3. for 문**

```
for 변수 in range(범위):
  반복할 문장
  ...
```

````python
for i in range(0, 10, 1): #start(이상), stop(미만), step
  print(i, end=' ')
````

````python
for i in range(10): #시작인덱스 기본값 : 0, 증가값 기본값 : 1
    print(i, end=' ')
````

````python
for i in range(0, 10) : # 증가값 기본값 : 1
    print(i, end=' ')
````

````python
len = range(0, 10, 1)
print(len)

for i in len:
  print(i, end=' ')
````

````python
# 1 ~ 100까지의 총합을 구하는 프로그램

sum = 0 # 합계를 저장할 변수

for i in range(1,101,1):
  sum += i
print(sum)
````

### **문제.**
score 리스트에 저장된 점수가 60점 이상인 학생이 몇명인지 알아보는 프로그램을 작성해보자  
* score = [90, 30, 50, 60, 80, 70, 100, 40, 20, 10]

````python

score = [90, 30, 50, 60, 80, 70, 100, 40, 20, 10]

for i in score:
  if i >=60:
    print(i)
````

````python
score = [90, 30, 50, 60, 80, 70, 100, 40, 20, 10]

count = 0

for i in score:
  if i >= 60:
    count+= 1

print(f'60점 넘는 학생의 수는 {count}명입니다.')
````

# **6. 다중 반복문**

* for 또는 while문이 2개이상 겹쳐있는 문장

````python
for num1 in range(1,4):
  print(f'현재 num1 : {num1}')
  for num2  in range(1,6):
    print(f'❤️')
````

````python
#'❤️' 를 이용하여 아래와 같은 도형을 만들어보자

#❤️❤️❤️❤️❤️  
#❤️❤️❤️❤️❤️
#❤️❤️❤️❤️❤️
#❤️❤️❤️❤️❤️
#❤️❤️❤️❤️❤️

for i in range(5):
  for j in range(5):
    print('❤️',end=" ")
  print()
````
