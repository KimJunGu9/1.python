# **1. 람다 함수**

* 단일문으로 표현되는 익명함수
* 코드상에서 한번만 사용되는 기능이 있을 때, 굳이 함수로 만들지 않고 일회성으로 만들어 사용할 때
* 함수 또한 클래스를 통해 생성된 객체

```python

def square1(x):
  return x ** 2

print(square1(10))
```

```python
# def 함수명(변수): 
#   return 값

# =

# lambda 변수 : 값
```

```python
square2 = lambda x: x ** 2
print(square2(7))
```

```python
def myKey(str):
  return len(str.strip())  # strip : 양쪽공백날리는것
```

```python
target = ['   apple', ' banana', 'orange    ', 'melon']
print(sorted(target, key=myKey))
```

```python
target = ['   app', ' bana', 'orange    ', 'melon']
print(sorted(target,key=myKey))    # 첫번째 줄에 있는 target을 두번째 줄에 있는 target에 넣고, key를 sort한다.     # sort는 오름차순정렬
```

```python
print(sorted(target,key=lambda x: len(x.strip())))
```

```python
add2 = lambda x, y: x * y

print(add2(10, 5))
```

# **2. filter 함수**

* 람다가 유용하게 사용되는 대표적인 함수
* 특정 조건을 만족하는 요소만 남기고 필터링

```python

nums = [1,2,3,4,6,9,10,14,15,18,20]

print(list(filter(lambda n:n %2 == 0, nums)))
```

# **3. map 함수**

* 함수형 프로그래밍의 기본 요소
* 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트를 반환

```python

nums = [1,2,3,4,6,9,10,14,15,18,20]

print(list(map(lambda n: n ** 2, nums)))
```

```python
nums = [1,2,3,4,6,9,10,14,15,18,20]

print(list(filter(lambda n : n % 2 == 0 , map(lambda n: n ** 2, nums))))
```
